__Exercise 1.1__ (10 pts - 2 each)

Recall that surrogate losses for large margin classification take the form, $\phi(y_i x_i^\top \beta)$ where $y_i \in \{-1,1\}$ and $\beta, x_i \in \mathbb R^p$.

The following functions are used as surrogate losses for large margin classification.  Demonstrate if they are convex or not, and follow the instructions.

1. exponential loss: $\phi(x) = e^{-x}$
1. truncated quadratic loss: $\phi(x) = (\max\{1-x,0\})^2$
1. hinge loss: $\phi(x) = \max\{1-x,0\}$
1. sigmoid loss: $\phi(x) = 1 - \tanh(\kappa x)$, for fixed $\kappa > 0$
1. Plot these as a function of $x$.

(This problem is due to notes of Larry Wasserman.) 


```python
import matplotlib.pyplot as plt
import numpy as np
import math
import pandas as pd
%matplotlib inline

variable = np.arange(-3,3,.1)
f1 = [math.exp(-x) for x in variable]
f2 = [max(1-x,0)**2 for x in variable]
f3 = [max(1-x,0) for x in variable]
k = 1
f4 = [1- math.tanh(k*x) for x in variable]

plt.figure(figsize=(20,12.5))

plt.subplot(221)
plt.plot(variable,f1)
plt.title('Exponential loss')

plt.subplot(222)
plt.plot(variable,f2)
plt.title('Truncated quadratic loss')

plt.subplot(223)
plt.plot(variable,f3)
plt.title('Hinge loss')

plt.subplot(224)
plt.plot(variable,f4)
plt.title('Sigmoid loss')

plt.show()
```

As we see from the graphs above, exponential, truncated quadratic and hinge losses are convex, but sigmoid loss is not.

__Exercise 1.2__ (10 pts)

Consider the truncated quadratic loss from (1.1.2).  For brevity let $a_+ = max\{a,0\}$ denote the positive part of $a$.

$$\ell(y_i,x_i,\beta) = \phi(y_i x_i^\top \beta) = (1-y_i x_i^\top \beta)_+^2$$

1. Consider the empirical risk, $R_n$ (the average loss over a training set) for the truncated quadratic loss.  What is gradient of $R_n$ in $\beta$?  Does it always exists?
1. Demonstrate that the gradient does not have continuous derivative everywhere.
1. Recall that support vector machines used the hinge loss $(1 - y_i x_i^\top)_+$ with a ridge regularization.  Write the regularized optimization method for the truncated quadratic loss, and derive the gradient of the regularized empirical risk.
1. Because the loss does not have continuous Hessian, instead of the Newton method, we will use a quasi-Newton method that replaces the Hessian with a quasi-Hessian (another matrix that is meant to approximate the Hessian).  Consider the following quasi-Hessian of the regularized objective to be $$G(\beta) = \frac 1n \sum_i 2 (x_i x_i^\top 1\{ y_i x_i^\top \beta > 1 \}) + 2 \lambda.$$  Demonstrate that the quasi-Hessian is positive definite, and write pseudo-code for quasi-Newton optimization.  (There was a correction in the lectures, that when minimizing a function you should subtract the gradient $\beta \gets \beta - H^{-1} g$).

1. The function $\phi(t) = (1-t)^2_+$ has continuous derivative $\phi'(t) = -2(1-t)_+$, and so the gradient of $R_n$ is
$$ \nabla_\beta R_n = \frac{1}{n} \sum \nabla_\beta (1 - y_i x_i^T \beta)^2_+ = \frac{1}{n} \sum -2(1 - y_i x_i^T \beta)_+\ y_i x_i,$$
which exists for all $\beta$.

2. The derivative of the function $\phi'(t) = -2(1-t)_+$ is equal to 0 for $t > 1$ and is equal to 2 for $t<1$. Thus, the derivative has discontinuity at $t=1$. Similarly, the derivative with respect to $\beta$ of the function $\nabla_\beta \ell(y_i, x_i,\beta) = -2(1 - y_i x_i^T \beta)_+\ y_i x_i$ in the direction of $x_i$ is discontinuous at a point $\beta_0$ with $x_i \beta_0 = y_i$.

3. Optimization method for truncated quadratic loss with ridge penalty is
$$\hat\beta = \mathrm{argmin} \left(R_n^r\right) = \mathrm{argmin} \left(\frac{1}{n}\sum (1 - y_i x_i^T \beta)^2_+ + \lambda \|\beta\|_2^2\right).$$
The gradient of regularized loss $R_n^r$ is
$$\nabla_\beta R_n^r = \left( \frac{1}{n} \sum -2(1 - y_i x_i^T \beta)_+\ y_i x_i \right) + 2 \lambda\beta.$$

4. Positive definite: given any vector $z$, we show that $z^T G(\beta) z > 0$:
$$z^T G(\beta) z = \frac{1}{n} \sum_i 2\ \mathrm{I}\{ y_i x_i^T \beta >1\} (z^T x_i)^2 + 2\lambda > 0,$$
since $\lambda>0$.
Pseudo-code for quasi-Newton optimization:
    1. Initialize $\beta$ randomly.
    2. Calculate gradient $g(\beta) = \nabla_\beta R_n^r = \left( \frac{1}{n} \sum -2(1 - y_i x_i^T \beta)_+\ y_i x_i \right) + 2 \lambda\beta.$
    3. Calculate quasi-Hessian $G(\beta) = \frac 1n \sum_i 2 (x_i x_i^\top 1\{ y_i x_i^\top \beta > 1 \}) + 2 \lambda.$
    4. Update $\beta$ with $\beta \leftarrow \beta - G(\beta)^{-1} g(\beta)$.
    5. Repeat steps (B.-D.) $N$ times until $\| \beta_N - \beta_{N-1}\| < \varepsilon$.
    
 Since the optimization problem is convex with positive-definite Hessian, the alrogithm will converge to a unique solution $\hat\beta$.

# HW3 Logistic, LDA, SVM

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

import sklearn.linear_model as skl_lm
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, classification_report, precision_recall_curve, roc_curve
from sklearn import preprocessing
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
# dataset path
data_dir = "."
```

The following code reads the data, subselects the $y$ and $X$ variables, and makes a training and test split.  This is the Abalone dataset and we will be predicting the age. V9 is age, 1 represents old, 0 represents young. 

```python
sample_data = pd.read_csv(data_dir+"/hw3.csv", delimiter=',')
sample_data.V1=sample_data.V1.factorize()[0]

X = np.array(sample_data.iloc[:,range(0,8)])
m = len(X[:,0])
X = np.append(X,np.ones((m,1)),1)
y = np.array(sample_data.iloc[:,8])
y = y*2 -1


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
```

__Exercise 2.1__ (10 pts) Perform logistic regression using Newton conjugate gradient.  You should save the predicted probabilities, and save the roc and pr curves (using roc_curve and precision_recall_curve) computed using the test set.

```python
def logit_matrices(beta , X, y):
    # Computing gradiend and Hessian for Newton's method.
    n = len(X[:,0])
    m = len(X[0,:])
    g = np.zeros(m)
    H = np.zeros((m,m))
    for i in range(n):
        inner = np.dot(beta, X[i,:])
        g += (-y[i] + math.tanh(inner))*X[i,:]
        H += (1 - math.tanh(inner)**2) * np.dot(np.array([X[i,:]]).T, np.array([X[i,:]]))
    g = g/n
    H = H/n
    return g,H

m = len(X_train[0,:])

epsilon = 0.01
N=0
beta_old = np.zeros(m)
beta_new = np.ones(m)/6
while np.linalg.norm(beta_new - beta_old) > epsilon and N<100:
    beta_old = beta_new
    (g,H) = logit_matrices(beta_old, X_train, y_train)
    beta_new = beta_old - np.dot(np.linalg.inv(H),g)
    N += 1

beta_hat = beta_new
l = len(X_test[:,0])
prob_logit = np.zeros(l)
for i in range(l):
    inner = np.dot(beta_hat, X_test[i,:])
    prob_logit[i] = np.exp(inner)/ (np.exp(-inner) + np.exp(inner))
```

```python
(fpr_logit,tpr_logit,thresholds) = roc_curve(y_test, prob_logit)
(prec_logit,rec_logit,thresholds) = precision_recall_curve(y_test, prob_logit)
```

__Exercise 2.2__ (10 pts) Do the same for linear discriminant analysis.

```python
X_train = X_train[:,:8]
X_test = X_test[:,:8]
X_train_1 = X_train[y_train==1]
X_train_0 = X_train[y_train==-1]

ave_0 = np.mean(X_train_0,0)
ave_1 = np.mean(X_train_1,0)

n,m = X_train.shape

sigma = np.zeros((m,m))

for x in X_train_0:
    sigma += np.dot(np.array([x-ave_0]).T, np.array([x-ave_0]))
for x in X_train_1:
    sigma += np.dot(np.array([x-ave_1]).T, np.array([x-ave_1]))
sigma = sigma/(n-2)

beta_disc = np.dot(np.linalg.inv(sigma),ave_1-ave_0)

l = len(X_test[:,0])
score_disc = np.zeros(l)
for i in range(l):
    score_disc[i] = np.dot(X_test[i,:],beta_disc)
```

```python
(fpr_disc,tpr_disc,thresholds) = roc_curve(y_test, score_disc)
(prec_disc,rec_disc,thresholds) = precision_recall_curve(y_test, score_disc)
```

__Exercise 2.3__ (10 pts) Do the same for support vector machines.

```python
clf = SVC(probability = True, kernel='linear')
clf.fit(X_train, y_train)

beta_svm = clf.coef_[0]

l = len(X_test[:,0])
score_svm = np.zeros(l)
for i in range(l):
    score_svm[i] = np.dot(X_test[i,:],beta_svm)

(fpr_svm,tpr_svm,thresholds) = roc_curve(y_test, score_svm)
(prec_svm,rec_svm,thresholds) = precision_recall_curve(y_test, score_svm)
```

__Exercise 2.4__ (10 pts) Plot and compare the ROC and PR curves for the above methods.

```python
import seaborn as sn

plt.figure(figsize = (15,7))

plt.subplot(121)
logit, = plt.plot(fpr_logit, tpr_logit, 'r',label = 'Logit')
disc, = plt.plot(fpr_disc, tpr_disc, 'b',label = 'Discriminant')
svm, = plt.plot(fpr_svm, tpr_svm, 'g', label = 'SVM')
plt.legend(handles=[logit,disc,svm])

plt.subplot(122)
logit, = plt.plot(prec_logit, rec_logit, 'r',label = 'Logit')
disc, = plt.plot(prec_disc, rec_disc, 'b',label = 'Discriminant')
svm, = plt.plot(prec_svm, rec_svm, 'g', label = 'SVM')
plt.legend(handles=[logit,disc,svm])

plt.show()
```
